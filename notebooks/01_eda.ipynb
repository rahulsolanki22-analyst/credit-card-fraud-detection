```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/creditcard.csv")
df.head()
```

```python
df['Class'].value_counts()
```

```python
sns.countplot(x='Class',data=df)
plt.title("Class Distribution (0 = Normal , 1 = Fraud)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Amount'], bins=50)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Amount vs Class")
plt.show()
```

```python
corr=df.corr()['Class'].sort_values()
corr
```

The dataset is clean with no missing values but is highly imbalanced. Fraud cases are extremely rare, making standard accuracy misleading. Transaction amounts are skewed and require scaling. Proper preprocessing and imbalance handling techniques are required before model training.
